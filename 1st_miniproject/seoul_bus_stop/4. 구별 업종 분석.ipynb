#### [4.2.2] 원하는 데이터 형태로 저장하기

* head, tail 등을 살펴본 결과를 토대로 header 값을 적절히 지정하여 데이터를 다시 불러오고 필요한 컬럼만 저장하세요

```python
#read_csv 재실행. 단, header값을 적절히 지정

seoul_business=pd.read_csv("4. business_type.csv",header=[1,2],sep=',',thousands=',')
seoul_business
```

```python
seoul_business.columns
```

```python
# 다시 불러온 데이터프레임에서 여러분 각자가 사용하고자 하는 컬럼만 가져오세요.
seoul_business.columns=['자치구','동','택시운송업_사업체수','택시운송업_종사자수','한식 일반 음식점업_사업체수','한식 일반 음식점업_종사자수',
                        '용달 화물자동차 운송업_사업체수','용달 화물자동차 운송업_종사자수','부동산 중개 및 대리업_사업체수','부동산 중개 및 대리업_종사자수',
                        '두발미용업_사업체수','두발미용업_종사자수','커피전문점_사업체수','커피전문점_종사자수',
                        '남녀용 겉옷 및 셔츠 도매업_사업체수','남녀용 겉옷 및 셔츠 도매업_종사자수','기타주점업_사업체수','기타주점업_종사자수',
                        '일반 교과 학원_사업체수','일반 교과 학원_종사자수','한식 육류요리 전문점_사업체수','한식 육류요리 전문점_종사자수']
seoul_business.head()
```

```python
# 여러분 각자가 사용하고자 하는 형태로 데이터를 정리합니다. 
# 예를 들어, 동 컬럼의 값이 소계인 항목들만 골라내 보겠습니다. 
# 위와 같이 동 컬럼의 소계를 구하면 자치구 별 값이 되므로 이후에 동 컬럼은 삭제하겠습니다. 결과적으로 자치구 별 값을 얻게 됩니다.

new_df=seoul_business.loc[seoul_business['동']=='소계']
new_df.drop('동',axis=1,inplace=True)
new_df.head()
```

```python
# 위에서 여러분 각자가 사용하고자 하는 형태로 정리한 해당 데이터프레임을 csv 파일('df_seoul_business.csv')로 저장하세요.

new_df.to_csv('df_seoul_business.csv')
```

```python
# 시각화 한글폰트 설정
import seaborn as sns
import matplotlib.pyplot as plt

plt.rc('font', family='Malgun Gothic')
sns.set(font="Malgun Gothic",#"NanumGothicCoding", 
        rc={"axes.unicode_minus":False}, # 마이너스 부호 깨짐 현상 해결
        style='darkgrid')        
```

#### [4.3.1] 데이터 분포 알아보기_1
* '자치구' 등 여러 변수를 기준으로 시각화 해보고 인사이트를 도출해보세요.

```python
# 아래에 코드를 작성하고 결과를 확인합니다.
df=pd.read_csv('df_seoul_business.csv',index_col=0,thousands=',')
df=df.reset_index(drop=True)
df
```

```python
plt.figure(figsize=(20,6))

plt.plot(df['자치구'],df['택시운송업_사업체수'],label='택시운송업')
plt.plot(df['자치구'],df['한식 일반 음식점업_사업체수'],label='한식 일반 음식점업')
plt.plot(df['자치구'],df['용달 화물자동차 운송업_사업체수'],label='용달 화물자동차 운송업')
plt.plot(df['자치구'],df['부동산 중개 및 대리업_사업체수'],label='부동산 중개 및 대리업')
plt.plot(df['자치구'],df['두발미용업_사업체수'],label='두발미용')
plt.plot(df['자치구'],df['커피전문점_사업체수'],label='커피전문점')
plt.plot(df['자치구'],df['남녀용 겉옷 및 셔츠 도매업_사업체수'],label='남녀용 겉옷 및 셔츠 도매업')
plt.plot(df['자치구'],df['기타주점업_사업체수'],label='기타주점업')
plt.plot(df['자치구'],df['일반 교과 학원_사업체수'],label='일반 교과 학원')
plt.plot(df['자치구'],df['한식 육류요리 전문점_사업체수'],label='한식 육류요리 전문점')

plt.xticks(rotation=45)
plt.legend(loc='best')
plt.show()
```

```python
selected_columns = ['택시운송업_종사자수', '한식 일반 음식점업_종사자수', '용달 화물자동차 운송업_종사자수', '부동산 중개 및 대리업_종사자수', '두발미용업_종사자수']
df[selected_columns].boxplot()
plt.title('주요 업종 종사자수 분포')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
