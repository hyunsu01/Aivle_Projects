# (실습준비) 데이터 불러오기부터

* 17년도부터 21년도 까지의 데이터를 train set, 22년도의 6개월치 데이터를 test set 으로 설정합니다.

```python
import pandas as pd

# 학습 데이터 로딩
df_total = pd.read_csv('df_total.csv')
df_test = pd.read_csv('df_test.csv')
```

# 1. 데이터 전처리

#### [실습문제1] 열 추가 
* 기존 '기준일ID'열에서 날짜 데이터를 추출하여 새로운 열로 추가 합니다.
    - pd.to_datetime() 함수를 사용해 '기준일ID'열을 datetime 형식으로 변경하겠습니다.
    - year, month, day 열로 기존 데이터 프레임에 연월일 정보를 저장합니다.
    - 데이터 프레임 열에 대해 날짜 정보를 얻을 때에는 **dt.year**, **dt.month**, **dt.day** 속성을 사용합니다. 

```python
# 아래에 실습코드를 작성하고 결과를 확인합니다.
df_total['기준일ID'] = pd.to_datetime(df_total['기준일ID'], format='%Y-%m-%d')
df_total['year'] = df_total['기준일ID'].dt.year
df_total['month'] = df_total['기준일ID'].dt.month
df_total['day'] = df_total['기준일ID'].dt.day

df_total.head()
```

```python
# '연도', '월', '일' 정보를 새로운 열로 추가
df_test['기준일ID'] = pd.to_datetime(df_test['기준일ID'], format='%Y-%m-%d')
df_test['year'] = df_test['기준일ID'].dt.year
df_test['month'] = df_test['기준일ID'].dt.month
df_test['day'] = df_test['기준일ID'].dt.day

df_test.head()
```

#### [실습문제2] 불필요한 열 제거 및 열 이름 변경
* df_total과 df_test에서 '총생활인구수'외에 불필요한 열을 제거 하도록 하겠습니다.
* 시간 데이터를 남기고, '시간대구분' column 을 'hour'로 rename 하도록 하겠습니다. 

```python
# 불필요한 열 제거 및 열 이름 변경
df_total = df_total[['기준일ID', '시간대구분', '총생활인구수', 'year', 'month', 'day']]
df_total = df_total.rename(columns={'시간대구분': 'hour'})
df_total.head()
```

```python
# 불필요한 열 제거 및 열 이름 변경
df_test = df_test[['기준일ID', '시간대구분', '총생활인구수', 'year', 'month', 'day']]
df_test = df_test.rename(columns={'시간대구분': 'hour'})
df_test.head()
```

```python
print(df_total.columns)
print(df_test.columns)
```

#### [실습문제3] 시프트 & 롤링
* shift()를 활용하여 이동할 행 수를 지정합니다.
* rolling()을 활용하여 추가할 이동평균 수를 지정합니다.

```python
# 하루 24시간
# 최근 3시간
# 바로 직전 시간
# 출퇴근시간대 (오전 6시-9시, 오후6시-9시) 인지 아닌지
# 공휴일, 휴일인지 아닌지
# 요일(주말, 주중) 여부
```

```python
# '기준일ID'를 인덱스로 설정
df_total.set_index('기준일ID', inplace=True)
df_test.set_index('기준일ID', inplace=True)
```

##### - 일별 '총생활인구수'의 평균

```python
# 일별 '총생활인구수'의 평균
df_total['daily_avg'] = df_total['총생활인구수'].rolling(window=24).mean()
df_test['daily_avg'] = df_test['총생활인구수'].rolling(window=24).mean()
```

```python
df_total.head()
```

```python
df_test.head()
```

##### - 최근 3시간 '총생활인구수'의 평균

```python
# 최근 3시간 '총생활인구수'의 평균
df_total['3hour_avg'] = df_total['총생활인구수'].rolling(window=3).mean()
df_test['3hour_avg'] = df_test['총생활인구수'].rolling(window=3).mean()
```

##### - 바로 직전 시간 '총생활인구수'

```python
# 바로 직전 시간 '총생활인구수'
df_total['1daybefore'] = df_total['총생활인구수'].shift(-1)
df_test['1daybefore'] = df_test['총생활인구수'].shift(-1)
```

##### - 출퇴근시간대 (오전 6시-9시, 오후6시-9시) 여부

```python
# 출퇴근시간대
df_total['출퇴근시간대'] = df_total['hour'].apply(lambda x: '출퇴근시간' if (6 <= x <= 9) or (18 <= x <= 21) else '아님')
df_test['출퇴근시간대'] = df_test['hour'].apply(lambda x: '출퇴근시간' if (6 <= x <= 9) or (18 <= x <= 21) else '아님')
```

```python
# 공휴일 데이터프레임 만들기
import pandas as pd
from workalendar.asia import SouthKorea

date_range = pd.date_range(start='2017-01-01', end='2022-12-31')

cal = SouthKorea()

holidays = []
for year in range(2017, 2023):
    year_holidays = cal.holidays(year)
    holidays.extend(year_holidays)

# 휴무일 데이터프레임 생성
holidays_df = pd.DataFrame(holidays, columns=['기준일ID', 'Holiday_Name'])

# 휴무일 데이터프레임 출력
print(holidays_df)
```

```python
holidays_df.info()
```

```python
# 휴무일 날짜 datatime으로 변경
holidays_df['기준일ID'] = pd.to_datetime(holidays_df['기준일ID'])
```

```python
holidays_df.info()
```

```python
# 기존 데이터에 휴무일 정보 결합 + 휴무일이 아닌 날짜는 0으로 저장

df_total = df_total.merge(holidays_df, how='left', on='기준일ID')
df_total['Holiday_Name'].fillna(0, inplace=True)
```

```python
df_test = df_test.merge(holidays_df, how='left', on='기준일ID')
df_test['Holiday_Name'].fillna(0, inplace=True)
```

#### [실습문제4] 결측치 처리 및 불필요한 열 제거
* 롤링 & 시프트 과정에서 발생한 결측치를 처리하도록 하겠습니다.
* 불필요한 열을 추가로 제거 하겠습니다.(시간 관련 행들)

```python
# 결측치 여부 확인
df_total.isnull().sum()
```

```python
# 결측치 여부 확인
df_test.isnull().sum()
```

```python
# 결측치 처리
df_total['daily_avg'].fillna(df_total['daily_avg'].mean(), inplace=True) # 평균값
df_total['3hour_avg'].interpolate(method='linear', inplace=True) # 선형보간법
df_total['1daybefore'].fillna(df_total['1daybefore'].mean(), inplace=True) # 평균값

df_test['daily_avg'].fillna(df_test['daily_avg'].mean(), inplace=True) # 평균값
df_test['3hour_avg'].interpolate(method='linear', inplace=True) # 선형보간법
df_test['1daybefore'].fillna(df_test['1daybefore'].mean(), inplace=True) # 평균값
```

```python
df_total['3hour_avg'].fillna(df_total['3hour_avg'].mean(), inplace=True) # 평균값
df_test['3hour_avg'].fillna(df_test['3hour_avg'].mean(), inplace=True) # 평균값
```

#### [실습문제5] train, test 데이터 나누기
* df_total과 df_test를 각각 train_x, train_y, test_x, test_y 로 나누어 주세요.
* 인덱스를 리셋 해주세요.

```python
# train_x, train_y 분리
train_x = df_total.drop(columns=['총생활인구수'])
train_y = df_total['총생활인구수']
test_x = df_test.drop(columns=['총생활인구수'])
test_y = df_test['총생활인구수']

# 인덱스 리셋
train_x = train_x.reset_index(drop=True)
train_y = train_y.reset_index(drop=True)
test_x = test_x.reset_index(drop=True)
test_y = test_y.reset_index(drop=True)
```
